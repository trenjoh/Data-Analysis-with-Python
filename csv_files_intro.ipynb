```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/weather_data.csv',header=0)
print(df)

```

```python
#deliting some rows
df.drop([24,25], inplace=True)
print(df)
```

```python
weights = np.array([3,4])
#results = df.dot(weights)
results = df @ weights

```

```python
climate_result = pd.concat([df,pd.DataFrame(results)], axis =1)
climate_result.columns = [*climate_result.columns[:-1], 'Results']
```
