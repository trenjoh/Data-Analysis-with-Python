# ANALYZING DATA FROM DATAFRAMES

```python
import pandas as pd
```

```python
dt = pd.read_csv('/content/drive/My Drive/Colab Notebooks/weather_data_updated.csv',header=0)
print(dt)
```

```python
Total_towns = dt['Town'].count()
print(  f'The total numbers of towns {Total_towns}' )
```

```python
Average_min_temp = dt["Min_temp(°C)"].mean()
print(f'The average minimum temperature is {Average_min_temp}')
```

```python
Average_max_temp =dt['Max_temp(°C)'].mean()
print(f"The Average maximum temperature is {Average_max_temp:.1f}")
```

```python
Total_min_temp = dt['Min_temp(°C)'].sum()
print(f'The total minimum temperature is {Total_min_temp}')
```

```python
Max_min_temp = dt['Min_temp(°C)'].max()
print(f'The max minimum temperature is {Max_min_temp}')
```

```python
High_temperature_rec = dt[dt['Max_temp(°C)'] >30]
print(High_temperature_rec)
```

```python
from IPython.display import display
with pd.option_context('display.max_rows', 6):
  display(dt[dt['Max_temp(°C)']>29])
```

# Displaying the grouped and aggregated dataframe


```python
# Grouping by 'Temperature Category' and calculate the mean of 'Min_temp(°C)' and 'Max_temp(°C)'
grouped_dt = dt.groupby('Temperature Category').agg({
    'Min_temp(°C)': 'mean',
    'Max_temp(°C)': 'mean'
}).round(2)

# Display the grouped and aggregated dataframe
print(grouped_dt)
```

```python
# Sort by 'Min_temp(°C)' in ascending order
sorted_dt_min = dt.sort_values(by='Min_temp(°C)', ascending=True)
sorted_dt_min
```

# MERGING DIFFERNT DATA FRAMES.


```python
# Dataframe 1 (df1)
data1 = {
    'Town': ['Nairobi', 'Mombasa', 'Kisumu', 'Nakuru'],
    'Min_temp(°C)': [14.0, 22.0, 20.0, 14.0],
    'Max_temp(°C)': [22.0, 29.0, 26.0, 20.0]
}
df1 = pd.DataFrame(data1)

# Dataframe 2 (df2)
data2 = {
    'Town': ['Nairobi', 'Mombasa', 'Eldoret', 'Nyeri'],
    'Population': [4500000, 1200000, 290000, 300000],
    'Elevation(m)': [1795, 50, 2100, 1780]
}
df2 = pd.DataFrame(data2)

# Merging the dataframes on 'Town' column using an inner join
merged_df = pd.merge(df1, df2, on='Town', how='inner')

# Display the merged dataframe
print(merged_df)

```

```python
merged_df = pd.merge(df1, df2, on='Town', how='left')
```
