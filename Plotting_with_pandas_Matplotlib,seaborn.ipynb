# MATPLOTLIB AND SEABORN

1) Line graphs-matplotlib



```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
# sns.lineplot(data=dt)
plt.plot(dt['Town'], dt['Min_temp(°C)'], marker = '*', label = 'min temperature', color = 'green', linewidth=3)
plt.plot(dt['Town'], dt['Max_temp(°C)'], marker = '*', label = 'min temperature', color = 'red', linewidth=3)
# Customizing
plt.title('Temperature in towns', fontsize=16)
plt.xlabel('Town',color='blue', fontsize=16)
plt.xticks(rotation=90)
plt.ylabel('Temperature', color='blue', fontsize=16)
plt.legend()
plt.show()
```

-seabon

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn style
sns.set(style='whitegrid')

plt.figure(figsize=(10,6))
sns.lineplot(x='Town', y='Min_temp(°C)', data=dt, marker='o', label='Min Temperature (°C)', color='blue')
sns.lineplot(x='Town', y='Max_temp(°C)', data=dt, marker='o', label='Max Temperature (°C)', color='orange')

# Customization
plt.title('Min and Max Temperatures by Town')
plt.xticks(rotation=90)
plt.ylabel('Temperature (°C)')
plt.legend()

plt.show()
```

# SCATTER PLOTS
- Matplotlib


```python
plt.figure(figsize=(15,5))
plt.scatter(dt['Town'], dt['Min_temp(°C)'], marker = '*', label = 'min temperature', color = 'red')
plt.title('Temperature in towns', fontsize=16)
plt.xlabel('Town',color='blue', fontsize=16)
plt.xticks(rotation=90)
plt.ylabel('Temperature', color='blue', fontsize=16)
plt.legend()
```

.Seaborn

```python
plt.figure(figsize=(13,5))

sns.scatterplot(x='Town', y='Min_temp(°C)', data=dt, color='red')

# Customization
plt.title('Seaborn Scatter Plot: Min vs Max Temperature')
plt.xlabel('Town',fontsize = 18, color = 'blue')
plt.xticks(rotation=90)
plt.ylabel('Max Temperature (°C)',fontsize = 18, color = 'blue')
plt.grid(True)

plt.show()
```

# HISTOGRAM
Matplotlib


```python
plt.figure(figsize=(14, 5))
plt.hist(dt['Min_temp(°C)'], bins=10, color='blue', edgecolor='black')
plt.xlabel('Min Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Histogram of Min Temperature')
plt.grid(True)
plt.show()
```

Seaborn

```python
plt.figure(figsize=(14,5))
sns.histplot(dt['Max_temp(°C)'], bins=5, color='purple', kde=True)
plt.title('Seaborn Histogram of Max Temperatures')
plt.xlabel('Max Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```

# HEATMAPS
Matplotlib


```python
import seaborn as sns
import numpy as np

plt.figure(figsize=(10,4))

corr = dt[['Min_temp(°C)', 'Max_temp(°C)']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap: Correlation between Min and Max Temperatures')
plt.show()
```

-seaborn


```python
plt.figure(figsize=(10, 4))
corr = dt[['Min_temp(°C)', 'Max_temp(°C)']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Seaborn Heatmap: Correlation between Min and Max Temperatures')
plt.show()
```
